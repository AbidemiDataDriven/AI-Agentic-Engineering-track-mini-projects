```python
from dotenv import load_dotenv 
from openai import OpenAI
import os 
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
```

```python
openai_api_key = os.getenv("OPENAI_API_KEY")
    
if openai_api_key:
    print(f"OpenAI API KEY Available {openai_api_key[:8]}")
else:
    print("Openaiapi key is not currently available")
```

```python
openai= OpenAI()
question = "Nigeria Government has been corrupted due to the poor infrastructural systems and incompentency leaders despites the natural resources, Hopefully Nigeria would get better soon with this new strategies and revolutionary systems"
messages = [{"role":"user", "content":question}]

response  = openai.chat.completions.create(
    model = "gpt-4o-mini",
    messages=messages
)
problem_solution= response.choices[0].message.content
print(problem_solution)
```

```python
display(Markdown(problem_solution))
```
